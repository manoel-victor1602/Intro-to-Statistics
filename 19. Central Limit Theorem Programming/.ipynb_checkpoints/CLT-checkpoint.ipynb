```python
%pylab inline
import random
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

# Programming Flips

```python
def flip(n):
    return [random.random() > 0.5 for i in range(n)]
```

```python
def flip2(n):
    for i in range(n):
        yield random.random()
```

```python
data = flip(1000)
data
```

```python
data = flip2(1000)
data
```

# Sets of Flips

```python
def sample(n):
    return [np.mean(flip(n)) for i in range(n)]
```

```python
outcomes = sample(1000)
plt.hist(outcomes, bins=30)
```
