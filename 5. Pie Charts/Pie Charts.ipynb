# Build a Chart

```python
parties = pd.Series([175000, 50000, 25000, 50000], index=['A', 'B', 'C', 'D'])
```

```python
parties.plot(kind='pie', subplots=True)
```

```python
parties['Total'] = parties.loc['A'] + \
                   parties.loc['B'] + \
                   parties.loc['C'] + \
                   parties.loc['D']  
```

```python
parties
```

```python
print(parties['A']/parties['Total'])
print(parties['B']/parties['Total'])
print(parties['C']/parties['Total'])
print(parties['D']/parties['Total'])
```

# Inferring Counts

```python
parties = pd.Series([240000], index=['Total'])
```

```python
parties['A'] = int(parties['Total'] * (7 / 12))
parties['B'] = int(parties['Total'] * (2 / 12))
parties['C'] = int(parties['Total'] * (1 / 12))
parties['D'] = int(parties['Total'] * (2 / 12))
```
